```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
# Preparing the dataset
texts = []
labels = []
label_names = [node["name"] for node in nodes]
```

```python
for node in nodes:
    texts.extend(node["labels"])
    labels.extend([node["name"]] * len(node["labels"]))
```

```python
print(texts)
print(labels)
```

```python
# Create multi-label binary matrix
binary_labels = np.array([[1 if name == lbl else 0 for name in label_names] for lbl in labels])
binary_labels
```

```python
# Model pipeline
# TfidfVectorizer optional...
# RandomForestClassifier ... but trying out different classifiers
pipeline = Pipeline([
    ("tfidf", CountVectorizer()),
    ("clf", MultiOutputClassifier(GradientBoostingClassifier()))
])
```

```python
# Train the model
pipeline.fit(texts, binary_labels)
```

```python
# Prediction
def predict_labels(input_text):
    preds = pipeline.predict([input_text])
    predicted_labels = [label_names[i] for i in range(len(preds[0])) if preds[0][i] == 1]
    return predicted_labels
```

```python
# pick a random text, and run prediction
i = np.random.randint(0, len(text_examples))
text = text_examples[i]["text"]
print(text)
print(predict_labels(text))
```

```python
threshold = 0.4
```

```python
# Convert labels to indices
label_indices = {name: idx for idx, name in enumerate(label_names)}
label_indices
```

```python
y = np.zeros((len(texts), len(label_names)))
for i, label in enumerate(labels):
    y[i, label_indices[label]] = 1
y
```

```python
# Model pipeline
pipeline = Pipeline([
    ("tfidf", CountVectorizer()),
    ("clf", RandomForestClassifier())
])
```

```python
# Train the model
pipeline.fit(texts, y)
```

```python
# Prediction with probabilities
def predict_labels_with_threshold(input_text, threshold=0.5):
    probabilities = pipeline.named_steps['clf'].predict_proba(
        pipeline.named_steps['tfidf'].transform([input_text])
    )
    probabilities = np.array([prob[:, 1] for prob in probabilities]).T
    predicted_labels = [
        label_names[i] for i in range(len(label_names)) if probabilities[0, i] >= threshold
    ]
    return predicted_labels
```

```python
print(predict_labels_with_threshold(text, threshold))
```
