```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import textwrap
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
nltk.download('stopwords')
```

```python
stops = set(stopwords.words('english')) # unique item list
```

```python
stops = stops.union({'said', 'would', 'could', 'told', 'also', 'one', 'two', 'mr', 'new', 'year'}) # extend the stop words list
```

```python
# Convert the set to a list
stops = list(stops)
```

```python
df = pd.read_csv('bbc_text_cls.csv')
df.head()
```

```python
vectorizer = TfidfVectorizer(stop_words=stops)
```

```python
X = vectorizer.fit_transform(df['text'])
```

```python
nmf = NMF(n_components=10, beta_loss='kullback-leibler', solver='mu', random_state=0)
```

```python
nmf.fit(X)
```

```python
def plot_top_words(model, feature_names, n_top_words = 10):

    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f'Topic {topic_idx +1}', fontdict={'fontsize': 30})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)
        fig.suptitle('LAD', fontsize=40)
```

```python
feature_names = vectorizer.get_feature_names_out()
plot_top_words(nmf, feature_names)
```

```python
Z = nmf.transform(X)
```

```python
# Pick a random document
# check which "topics" are associated with it
# Are they related to the true label?

np.random.seed(0)
i = np.random.choice(len(df))
z = Z[i]
print(df.iloc[i]['text'])
print(textwrap.fill(df.iloc[i]['text']))
topics = np.arange(10) + 1

fig, ax = plt.subplots()
ax.barh(topics, z)
ax.set_xlabel('Probability')
ax.set_yticks(topics)
ax.set_title('True label: %s' % df.iloc[i]['labels'])
```
